```python
import pandas as pd
import matplotlib.pyplot as plt

# import all csv datafiles
df_ice_ran_105 = pd.read_csv("csv result/iceland_network_results_beta_105.csv")
df_usa_ran_105 = pd.read_csv("csv result/US_network_results_beta_105.csv")
df_ice_tar_105 = pd.read_csv("csv result/iceland_network_beta_1.05_results.csv")
df_usa_tar_105 = pd.read_csv("csv result/US_network_beta_1.05_results.csv")
```

```python
df_ice_ran_105.head()
```

```python
# extract column of values from csv file
alpha = df_ice_ran_105["Alpha"]
ice_ran_105_dc = df_ice_ran_105 ["degree"]
ice_ran_105_bc = df_ice_ran_105 ["betweenness"]
ice_ran_105_cc = df_ice_ran_105 ["closeness"]

ice_tar_105_dc = df_ice_tar_105 ["degree"]
ice_tar_105_bc = df_ice_tar_105 ["betweenness"]
ice_tar_105_cc = df_ice_tar_105 ["closeness"]

usa_ran_105_dc = df_usa_ran_105 ["degree"]
usa_ran_105_bc = df_usa_ran_105 ["betweenness"]
usa_ran_105_cc = df_usa_ran_105 ["closeness"]

usa_tar_105_dc = df_usa_tar_105 ["degree"]
usa_tar_105_bc = df_usa_tar_105 ["betweenness"]
usa_tar_105_cc = df_usa_tar_105 ["closeness"]
```

```python
# plot the figures to compare the random failure and targeted attack on US network
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Degree Centrality Plot
axes[0].plot(alpha, usa_ran_105_dc, label="random failure", marker='o')
axes[0].plot(alpha, usa_tar_105_dc, label="target attack", marker='o')
axes[0].set_xlabel("Alpha")
axes[0].set_ylabel("Mean fraction of failed nodes (I)")
axes[0].set_title("Degree Centrality")
axes[0].grid()
axes[0].legend()

# Betweenness Centrality Plot
axes[1].plot(alpha, usa_ran_105_bc, label="random failure", marker='o')
axes[1].plot(alpha, usa_tar_105_bc, label="target attack", marker='o')
axes[1].set_xlabel("Alpha")
axes[1].set_title("Betweenness Centrality")
axes[1].grid()
axes[1].legend()

# Closeness Centrality Plot
axes[2].plot(alpha, usa_ran_105_cc, label="random failure", marker='o')
axes[2].plot(alpha, usa_tar_105_cc, label="target attack", marker='o')
axes[2].set_xlabel("Alpha")
axes[2].set_title("Closeness Centrality")
axes[2].grid()
axes[2].legend()

fig.suptitle("Comparison of Random Failure and Targeted Attack on US Network (Beta = 1.05)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# plot the figures to compare the random failure and targeted attack on iceland network
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Degree Centrality Plot
axes[0].plot(alpha, ice_ran_105_dc, label="random failure", marker='o')
axes[0].plot(alpha, ice_tar_105_dc, label="target attack", marker='o')
axes[0].set_xlabel("Alpha")
axes[0].set_ylabel("Mean fraction of failed nodes (I)")
axes[0].set_title("Degree Centrality")
axes[0].grid()
axes[0].legend()

# Betweenness Centrality Plot
axes[1].plot(alpha, ice_ran_105_bc, label="random failure", marker='o')
axes[1].plot(alpha, ice_tar_105_bc, label="target attack", marker='o')
axes[1].set_xlabel("Alpha")
axes[1].set_title("Betweenness Centrality")
axes[1].grid()
axes[1].legend()

# Closeness Centrality Plot
axes[2].plot(alpha, ice_ran_105_cc, label="random failure", marker='o')
axes[2].plot(alpha, ice_tar_105_cc, label="target attack", marker='o')
axes[2].set_xlabel("Alpha")
axes[2].set_title("Closeness Centrality")
axes[2].grid()
axes[2].legend()

fig.suptitle("Comparison of Random Failure and Targeted Attack on Iceland Network (Beta = 1.05)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# calculate the difference between the random failure and target attack
diff_ice_dc_105 = abs(ice_ran_105_dc - ice_tar_105_dc)
diff_ice_bc_105 = abs(ice_ran_105_bc - ice_tar_105_bc)
diff_ice_cc_105 = abs(ice_ran_105_cc - ice_tar_105_cc)

diff_usa_dc_105 = abs(usa_ran_105_dc - usa_tar_105_dc)
diff_usa_bc_105 = abs(usa_ran_105_bc - usa_tar_105_bc)
diff_usa_cc_105 = abs(usa_ran_105_cc - usa_tar_105_cc)
```

```python
# plot the difference for the iceland network
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].plot(alpha, diff_ice_dc_105, marker='o')
axes[0].set_xlabel("Alpha")
axes[0].set_ylabel("Difference in mean fraction of failed nodes")
axes[0].set_title("Degree Centrality")
axes[0].grid()

axes[1].plot(alpha, diff_ice_bc_105, marker='o')
axes[1].set_xlabel("Alpha")
axes[1].set_ylabel("Difference in mean fraction of failed nodes")
axes[1].set_title("Betweenness Centrality")
axes[1].grid()

axes[2].plot(alpha, diff_ice_cc_105, marker='o')
axes[2].set_xlabel("Alpha")
axes[2].set_ylabel("Difference in mean fraction of failed nodes")
axes[2].set_title("Closeness Centrality")
axes[2].grid()

fig.suptitle("Comparison of Random Failure and Targeted Attack on Iceland Network (Beta = 1.05)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# plot the difference for the US network
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].plot(alpha, diff_usa_dc_105, marker='o')
axes[0].set_xlabel("Alpha")
axes[0].set_ylabel("Difference in mean fraction of failed nodes")
axes[0].set_title("Degree Centrality")
axes[0].grid()

axes[1].plot(alpha, diff_usa_bc_105, marker='o')
axes[1].set_xlabel("Alpha")
axes[1].set_ylabel("Difference in mean fraction of failed nodes")
axes[1].set_title("Betweenness Centrality")
axes[1].grid()

axes[2].plot(alpha, diff_usa_cc_105, marker='o')
axes[2].set_xlabel("Alpha")
axes[2].set_ylabel("Difference in mean fraction of failed nodes")
axes[2].set_title("Closeness Centrality")
axes[2].grid()

fig.suptitle("Comparison of Random Failure and Targeted Attack on US Network (Beta = 1.05)", fontsize=14)
plt.tight_layout()
plt.show()
```
